# Eqaution solve

# Ax = b
A -> Given Matrix

x -> Input vector

b -> Output vector

```python
A = np.array([[2,3,-1],
              [4,1,5],
              [3,0,6]])
              
b = np.array([[1],
              [2],
              [3]])

a,k,c = np.linalg.solve(A,b)

print(f"x = {a}, y = {k}, y = {c}")
```

# Trace
Trace is the sum of diagonal elements of a square matrix.

```python
# without using trace function
def trace(A):
    # check square R^n*n
    if A.shape[0] != A.shape[1]:
        return "Matrix must be n*n"
    
    diag_sum = 0
    for i in range(A.shape[0]):
        diag_sum += A[i, i]
    
    return diag_sum
        
A = np.array([[2,1,2],
              [1,9,2],
              [6,3,3]])
print(f"Trace of A = {trace(A)}")
```

```python
# using built-In-function
dia_sum = np.linalg.trace(A)
dia_sum
```
