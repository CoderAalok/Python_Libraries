# Eqaution solve

# Ax = b
A -> Given Matrix

x -> Input vector

b -> Output vector

```python
A = np.array([[2,3,-1],
              [4,1,5],
              [3,0,6]])
              
b = np.array([[1],
              [2],
              [3]])

a,k,c = np.linalg.solve(A,b)

print(f"x = {a}, y = {k}, y = {c}")
```

# Trace
Trace is the sum of diagonal elements of a square matrix.

```python
# without using trace function
A = np.array([[2,11,2],
              [1,4,2],
              [6,3,2]])

if A.shape[0] != A.shape[1]:
    print(None)

dia_sum = 0
for i in range(A.shape[0]):
    dia_sum += A[i,i]

dia_sum 
```

```python
# using built-In-function
dia_sum = np.linalg.trace(A)
dia_sum
```
