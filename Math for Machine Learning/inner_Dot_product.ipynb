# Manhattan distance
d=i=1∑n​∣xi​−yi​∣

```python
# general 2D Manhattan distance formula: d = |x1-x2| + |y1-y2| (when A = (x1,y1) and B = (x2,y2))
# for nD formula: d = |x1-y1| + |x2-y2| + ..... + |xn-yn|
# |x1-y1| here just x2 in place y1 but actually happening |x1-x2|+ |y1-y2|+...., just variable notation is different
# which looks like A = [x1,x2,...,xn], B = [y1,y2,...,yn]

#           [x1,x2] # here x1 = x1 and x2 = y1
A = np.array([2,3])

#           [y1,y2] # here y1 = x2 and y2 = y2
B = np.array([7,5])


md = np.sum(np.abs(A - B))
x = np.sum(np.abs(B))

print(f"Manhattan distance between A and B: {md}")
print(f"Manhattan norm of B: {x}")
```

# Euclidean distance

```python
X = np.array([4,2])
Y = np.array([5,1])

ed = np.sqrt(np.sum(np.abs(X - Y)**2))
y = np.sqrt(np.sum(np.abs(X)**2))
print(f"Euclidean distance between X and Y: {ed}")
print(f"Euclidean norm of X: {y}")
```

# Projection of a on b or b on a

It means the component of a along the direction of b

```python
a = np.array([5,3])
b = np.array([3,-2])

origin = np.zeros(a.shape)

len_ab = np.dot(a,b)/(np.linalg.norm(b))
len_ba = np.dot(a,b)/(np.linalg.norm(a))

proj_b_on_a = np.dot(a,b)/(np.linalg.norm(a))**2 * a

print('Magnitude of a and b is ', len_ab)
print(f"Projection of a on b = {proj_b_on_a}")
```

```python
fix, ax = plt.subplots(figsize=(10,7))

ax.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1, units='xy')
ax.quiver(*origin, *b,  angles='xy', scale_units='xy', scale=1, units='xy')
ax.quiver(*origin, *proj_b_on_a,  angles='xy', scale_units='xy', scale=1, units='xy')

# ax.annotate('Projected b', xy=(2.3,2), xytext=(0, 1), textcoords='axes fraction', va='bottom', ha='left', arrowprops=dict(facecolor='red', shrink=5))


ax.plot([b[0], proj_b_on_a[0]], [b[1], proj_b_on_a[1]], 'g--')

ax.text(a[0],a[1], 'a', color='r')
ax.text(b[0],b[1], 'b', color='r')


ax.grid()
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_aspect('equal')

plt.show()
```

# Projection of b on a

```python
a = np.array(([-2,4]))
b = np.array((3,5))
origin = np.zeros(2)

proj = a@b/np.linalg.norm(a)**2 * a  # projection b on a
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *b, angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(*origin, *proj, angles='xy', scale_units='xy', scale=1, color='y')
ax.plot([b[0], proj[0]], [b[1], proj[1]], 'g--')


ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid()

ax.text(a[0],a[1], 'b', color='k')
ax.text(b[0],b[1], 'a', color='r')

ax.set_aspect('equal')

plt.show()
```
