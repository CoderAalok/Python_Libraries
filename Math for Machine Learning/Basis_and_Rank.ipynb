```python
import numpy as np
import matplotlib.pyplot as plt
```

# # Basis

# Vector represented as standard way

```python
# before scale(changes of basis)
v = np.array([[3],[2]])

i = np.array([[1],[0]])
j = np.array([[0],[1]])

origin = np.zeros(2)
i
```

```python
fig, ax = plt.subplots()
ax.quiver(*origin, *v, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *i, color='r', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *j, color='g', angles='xy', scale_units='xy', scale=1)

# artist
ax.grid()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')

# text 
ax.text(*v[0], *v[1], s='v', color='k')
ax.text(*i[0], *i[1], s='i', color='r')
ax.text(*j[0], *j[1], s='j', color='g')

plt.show()
```

```python
# scale basis of vector
basis_i = 4*i
basis_j = 3 * j
```

```python

fig, ax = plt.subplots(figsize=(8,5))
ax.quiver(*origin, *v, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *basis_i, color='r', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *basis_j, color='g', angles='xy', scale_units='xy', scale=1)

ax.grid()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')

# text 
ax.text(*v[0], *v[1], s='v', color='k')
ax.text(*basis_i[0], *basis_i[1], s=f'{basis_i[0][0]}basis_i', color='r')
ax.text(*basis_j[0], *basis_j[1], s=f'{basis_j[1][0]}basis_j', color='g')

plt.show()
```

# Vector represented as different way

```python
# v = c1*b1 + c2*b2
# take a new different vector
v = np.array([-1,2])  # c1 -> -1, c2 -> 2

# (randomly choose) basis vector
b1 = np.array([2,1])
b2 = np.array([-1,1])
# expressed the new basis vector getting a vector
v = v[0]*b1 + v[1]*b2  

origin = np.zeros(2)
print(v) # vector 
```

```python
fig, ax = plt.subplots()
ax.quiver(*origin, *b1, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *b2, color='r', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *v, color='y', angles='xy', scale_units='xy', scale=1)

# figure
ax.grid()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')

# text
ax.text(b1[0],b1[1], s='b1', color='k')
ax.text(b2[0],b2[1], s='b2', color='r')
ax.text(v[0],v[1], s='v', color='g')


plt.show()
```

# # Rank

```python
# n*n matrix
A = np.array([[2,0,7],[0,1,0],[0,0,0]])

# Rank properties
rk = np.linalg.matrix_rank(A.T)
# rk == A.ndim
print(rk)

# m*n matrix
M = np.random.randint(0,9, size=((3,2)))
rk = np.linalg.matrix_rank(M)
print(M)
print(rk)
```
