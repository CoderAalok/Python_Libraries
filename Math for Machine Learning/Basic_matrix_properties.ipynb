# Matrix properties
1) Associativity
2) Distribuitivity

```python
# Associativity
A = np.array([[2,1,4],
              [5,1,5]]) # R^2*3

# Column of A is row of B
B = np.array([[2,5,2],
              [4,6,7],
              [8,2,4]]) # R^3*3

# And column of B is row of C
C = np.array([[4,2],
              [3,6],
              [9,1]]) # R^3*2

(A @ B) @ C == A @ (B @ C)
```

```python
# Distributivity
# For all A, B ∈ R^M*N
A = np.array([[2,7,4],
              [9,1,5]]) # R^2*3

B = np.array([[3,1,7],
              [7,8,2]])
#column of A is row of C; C ∈ R^N*P
C = np.array([[3,4],
             [1,6],
             [9,3]])  # R^3*2

(A + B) @ C == (A @ C) + (B @ C)
```

# Inverse Matrix
Inverse matrix does exist, if non-singular/invertible/regular ortherwise not.
det(A) != 0

```python
A = np.array([[3,3],
              [6,5]])
det = np.linalg.det(A)
print(np.allclose(det,-3))

# Calculate inverse
A_inv = np.linalg.inv(A)

# Identity matrix
I = np.eye(A.shape[0])

# A @ Inv == I
# Inv @ A == I

np.allclose(A @ A_inv, A_inv @ A)
A @ I


# print(f"Determinant of A = {np.round(det)}")
# print(f"Inverse of A =\n {A_inv}")
```
