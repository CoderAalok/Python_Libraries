```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Standard vectors
i = np.array([1,0])
j = np.array([0,1])

# these two vector now mapped to arbitrary vectors
vi = np.array([3,0])
vj = np.array([1,3])

# origin
origin = np.zeros(2)

# scaled by 2
si = 2*vi
sj = 2*vj
```

# Plotting into 2D
# 

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.quiver(*origin, *i, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *j, color='r', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *vi, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *vj, color='r',angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *si, color='g',angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *sj, color='y',angles='xy', scale_units='xy', scale=1)


# figure
ax.grid()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
ax.set_aspect('equal')

# text
ax.text(i[0], i[1], s='i', color='r')
ax.text(j[0], j[1], s='j', color='k')


plt.show()
```

# Find Eigenvector 
(A - kI)x = 0

A -> Given matrix

k -> scalar

I -> Identity matrix

x -> vector


```python
A = np.array([[4,2],[1,3]])

I = np.array([[1,0],
              [0,1]])
k = 5

x = np.array([2,1])

# matrix-vector multiplication
result = (A - (k*I)) @ x
result == 0 # this make ensure that eigenvalue k of matrix A
```

# Collinear/Collinearity and Codirected/Codirection
Collinear -> Two-vectors are collinear if vectors points in the same line or the opposite direction.

Codirection -> Two vectors that points in the same line.

```python
# Two-vectors
v1 = np.array([2,3])
v2 = np.array([4,6]) #this vector in same line

#this vector flip opposite direction
vi = np.array([2,3])
vj = np.array([-4,-6])

origin = np.zeros(2) 
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.quiver(*origin, *v1, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *v2, color='r', angles='xy', scale_units='xy', scale=1)

# axes lines
ax.axhline(0, color='k')
ax.axvline(0, color='k')

# figure
ax.grid()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_aspect('equal')

# text
ax.text(v1[0], v1[1], s='v1', color='k')
ax.text(v2[0], v2[1], s='v2', color='r')
ax.text(v2[0], 0.5*v2[1], s='Collinear\n and Codirection', color='k')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.quiver(*origin, *vi, color='b', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *vj, color='y', angles='xy', scale_units='xy', scale=1)



# axes lines
ax.axhline(0, color='k')
ax.axvline(0, color='k')

# figure
ax.grid()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_aspect('equal')

# text
# 
ax.text(vi[0], vi[1], s='vi', color='k')
ax.text(vj[0], vj[1], s='vj', color='r')
ax.text(vi[0], 2*vi[1], s='Collinear', color='k')

plt.show()
```

```python
x = np.linspace(1,-1, 30)
y = np.linspace(1,-1, 30)
x,y = np.meshgrid(x,y)

A = np.zeros((2,2))
theta = np.pi/4


A[0,0] = -np.cos(theta)
A[0,1] = np.sin(theta)
A[1,0] = np.sin(theta)
A[1,1] = np.cos(theta)

m = A[0,0]*x + A[0,1]*y
n = A[1,0]*x + A[1,1]*y
```

```python
fig,ax = plt.subplots(figsize=(6,6))

ax.grid()
ax.axhline(0, color='k')
ax.axvline(0, color='k')

ax.scatter(m,n, c=y)

plt.show()
```
