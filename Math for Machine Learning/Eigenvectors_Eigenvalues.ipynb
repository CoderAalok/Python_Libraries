# Calculates  Eigenvalues and Eigenvectors
 
(A - kI)x = 0, x != 0

so, for non-zero vector, a matrix must be singular

det(A-kI) = 0

A -> Square matrix

k -> Eigenvalue (scalar) 

I -> Identity matrix

x -> non-zero vector


```python
A = np.array([[4,2],
              [1,3]])

# standard basis vector
basis = np.eye(2)
A_basis = A @ basis

# Calculating eigenvalue and eigenvector (Original matrix)
lam, x = np.linalg.eig(A)
Ax = A @ x
lam_x = lam * x

# Calculating eigenvalue and eigenvector (Transpose of matrix)
val, y = np.linalg.eig(A.T)
AT_basis = A.T @ basis
ATy = A.T @ y
val_y = val * y

# Output
print(f"Check eigenvector:\n")
print(f"Ax = lam*x -> {np.allclose(Ax, lam_x)}")
print(f"ATy = val*y -> {np.allclose(ATy, val_y)}")

print(f"\nTranformation matrix Ax =\n {Ax}")
print(f"Eigenvalue of A =\n {np.diag(lam)}")
print(f"Eigenvector of A =\n {x}")

print(f"\nTransformation of matrix ATy  =\n {ATy}")
print(f"Eigenvalue of AT =\n {np.diag(val)}")
print(f"Eigenvector of AT =\n {y}")

print("\nNOTE: A and its transpose AT possess same eigenvalues but not necessary to be same eigenvectors.")
```

# Eigendecomposition
It diagonalized the matrices, off diagonal are zero values. Diagonal matrix powers are trivial.

Mathematically, A = Q Λ Q^−1

simplify above formula: Λ = Q A Q^-1  -> this gives diagonal matrix 

Q -> Eigenvectors as columns  (direction)

Λ -> Eigenvalues on diagonal  (scaling)

Q^-1 -> Inverse of Q

# For 2D

```python
# matrix = np.array([[2,4],[3,1]]) #original matrix
matrix = np.random.randint(1,10, size=((2,2)))

# Standard basis vector(x) to representing the co-ordinate system
# x = np.array([[1,0],[0,1]])
x = np.eye(2)
Ax = matrix @ x

# Eigendecompositon of a square matrix A
eig_val,Q = np.linalg.eig(matrix)
eig_val = np.diag(eig_val)
inv_Q = np.linalg.inv(Q)

# Reconstruct A
A = Q @ eig_val @ inv_Q

# Diagonalization
# Diagonal matrix power k
D = inv_Q @ (np.linalg.matrix_power(A,1)) @ Q

# print(f"Transformed of basis =\n {Ax}")
print(f"Diagonalized of matrix =\n {np.round(D)}")
# print(f"Eigenvector of matrix =\n {Q}")
# print(f"Eigenvalue of matrix =\n {np.round(eig_val)}")
# print(f"Inverse of Q =\n {inv_Q}")
# print(f"Eigendecomposition of matrix =\n {A}")
```

```python
origin = np.zeros(A.shape)
fig,ax = plt.subplots(figsize=(8,6))

ax.quiver(*origin, *x, color='r', angles='xy', scale_units='xy', scale=1, label='Basis vector')
ax.quiver(*origin, *Ax, color='g', angles='xy', scale_units='xy', scale=1, label='Matrix transforms with basis')
ax.quiver(*origin, *Q, color='b', angles='xy', scale_units='xy', scale=1, label="Eigenvector 'Q'")
ax.quiver(*origin, *D, color='m', angles='xy', scale_units='xy', scale=1, label='Diagonal matrix power')


ax.axhline(color='k')
ax.axvline(color='k')

ax.grid()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
ax.set_aspect("equal")
ax.legend()

plt.show()
```

# Power Iteration
This is usually use for finding eigenvalue for large matrices.

Steps are:

1) Starts with any random vector
2) Repeatedly multiply by matrix A (v -> Ax -> A^2x ->.....-> A^nv)
3) Keeps normalize to prevent from overflow
4) The vector converges to the fixed direction(dominant eigenvector)

```python
A = np.array([[2,1],[1,2]])
eig_val,eig_vec = np.linalg.eig(A)
x = np.array([3,4])

for _ in range(20):
    x = A@x/np.linalg.norm(A@x)

eig_val
```

# Connecting Eigendecomposition to SVD
It works for any matrices A E R^m*n.

A = U S V^T

V^T -> Input rotate

S -> Diagonal matrix of singular value

U -> Output rotate

```python
A = np.array([
    [1,3],
    [4,5]])

# # SVD Calculating data
U, S, VT = np.linalg.svd(A)

# Eigendecomposition
E, Q = np.linalg.eig(A)
inv_Q = np.linalg.inv(Q)
E = np.diag(E)

# # Diagonalize matrix
S = np.diag(S)

# transformation with standard basis vector
x = np.eye(2)
Ax = A @ x


# # Outputs
print("****Singular Value Decomposition****")
print(f"Eigenvalue of left singular vectors U =\n {U}")
print(f"\nEigenvalue of right singular vectors V =\n {VT}")
print(f"\nDiagonal matrix E =\n {S}")


print("\n*****Eigendecomposition*****")
print(f"Basis of eigenvector of A =\n {Q}")
print(f"\nDiagonal matrix =\n {E}")
print(f"\nInverse of Q =\n {inv_Q}")
```

# # Covariance Matrix

```python
def Cov(X):
    # No. of data
    N = X.shape[0]
    
    # check invalid case
    if N < 2 or X.ndim < 2:
        return None
    
    # Step 1: Calculate mean
    mu = np.mean(X, axis=0)

    # Step 2: Calculate Centered Matrix
    X_centered = X - mu

    # Step 3: Compute Covariance matrix
    cov = 1/(N-1) * (X_centered.T @ X_centered)

    # result
    return cov

X = np.array([[2,3],
              [4,5]])

# column-wise operation perform
# Raw 
print(Cov(X))
print('-------------------------')
# Built-in-function
print(np.cov(X, rowvar=False))


# Normalization
n = np.linalg.norm(Cov(X))
# After normalization row-wise operation perform
# Raw 
print(Cov(X)/n)
print('-------------------------')
# Built-in-function
print(np.cov(X))
```
