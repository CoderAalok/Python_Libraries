```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Standard vectors
i = np.array([1,0])
j = np.array([0,1])

# these two vector now mapped to arbitrary vectors
vi = np.array([3,0])
vj = np.array([1,3])

# origin
origin = np.zeros(2)

# scaled by 2
si = 2*vi
sj = 2*vj
```

# Plotting into 2D
# 

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.quiver(*origin, *i, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *j, color='r', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *vi, color='k', angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *vj, color='r',angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *si, color='g',angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *sj, color='y',angles='xy', scale_units='xy', scale=1)


# figure
ax.grid()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
ax.set_aspect('equal')

# text
ax.text(i[0], i[1], s='i', color='r')
ax.text(j[0], j[1], s='j', color='k')


plt.show()
```
