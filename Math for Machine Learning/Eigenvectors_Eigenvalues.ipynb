# Calculates  Eigenvalues and Eigenvectors
 
(A - kI)x = 0, x != 0

so, for non-zero vector, a matrix must be singular

det(A-kI) = 0

A -> Square matrix

k -> Eigenvalue (scalar) 

I -> Identity matrix

x -> non-zero vector


```python
A = np.array([[4,2],[1,3]])
# Matrix transformation

# Transpose matrix A
AT = A.T

# origin
origin = np.zeros(A.shape)

# standard basis vector
x = np.eye(2)
Ax = A @ x

# Calculating eigenvalue and eigenvector (Original matrix)
eigen_val, eigen_vec = np.linalg.eig(A)

# Calculating eigenvalue and eigenvector (Transpose of matrix)
val,vec = np.linalg.eig(AT)
ATx = AT @ x

# Output
print(f"Transpose of matrix A transform under the basis vector =\n {ATx}")
print(f"Matrix tranformation under the basis vector =\n {Ax}")
print(f"Eigenvalue of A = {eigen_val}")
print(f"Eigenvector of A =\n {eigen_vec}")
print(f"Eigenvalue of AT = {val}")
print(f"Eigenvector of AT =\n {vec}")
print("NOTE: A and its transpose AT possess same eigenvalues but not necessary to be same eigenvectors.")
```

```python
fig,ax = plt.subplots(figsize=(8,6))

ax.quiver(*origin, *x, color='r', angles="xy", scale_units='xy', scale=1, label='Basis vector')
ax.quiver(*origin, *Ax, color='g', angles="xy", scale_units='xy', scale=1, label='Standard basis (Ax)')
ax.quiver(*origin, *eigen_vec, color='b', angles="xy", scale_units='xy', scale=1, label='Eigenvector of (A)')

ax.grid()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')
ax.legend()

plt.show()
```

# Eigendecomposition
It diagonalized the matrices, off diagonal are zero values. Diagonal matrix powers are trivial.

Mathematically, A = Q Λ Q^−1

simplify above formula: Λ = Q A Q^-1  -> this gives diagonal matrix 

Q -> Eigenvectors as columns  (direction)

Λ -> Eigenvalues on diagonal  (scaling)

Q^-1 -> Inverse of Q

# For 2D

```python
# matrix = np.array([[2,4],[3,1]]) #original matrix
matrix = np.random.randint(1,10, size=((2,2)))

# Standard basis vector(x) to representing the co-ordinate system
# x = np.array([[1,0],[0,1]])
x = np.eye(2)
Ax = matrix @ x

# Eigendecompositon of a square matrix A
eig_val,Q = np.linalg.eig(matrix)
eig_val = np.diag(eig_val)
inv_Q = np.linalg.inv(Q)

# Reconstruct A
A = Q @ eig_val @ inv_Q

# Diagonalization
# Diagonal matrix power k
D = inv_Q @ (np.linalg.matrix_power(A,1)) @ Q

# print(f"Transformed of basis =\n {Ax}")
print(f"Diagonalized of matrix =\n {np.round(D)}")
# print(f"Eigenvector of matrix =\n {Q}")
# print(f"Eigenvalue of matrix =\n {np.round(eig_val)}")
# print(f"Inverse of Q =\n {inv_Q}")
# print(f"Eigendecomposition of matrix =\n {A}")
```

```python
origin = np.zeros(A.shape)
fig,ax = plt.subplots(figsize=(8,6))

ax.quiver(*origin, *x, color='r', angles='xy', scale_units='xy', scale=1, label='Basis vector')
ax.quiver(*origin, *Ax, color='g', angles='xy', scale_units='xy', scale=1, label='Matrix transforms with basis')
ax.quiver(*origin, *Q, color='b', angles='xy', scale_units='xy', scale=1, label="Eigenvector 'Q'")
ax.quiver(*origin, *D, color='m', angles='xy', scale_units='xy', scale=1, label='Diagonal matrix power')


ax.axhline(color='k')
ax.axvline(color='k')

ax.grid()
ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
ax.set_aspect("equal")
ax.legend()

plt.show()
```

# Power Iteration
This is usually use for finding eigenvalue for large matrices.

Steps are:

1) Starts with any random vector
2) Repeatedly multiply by matrix A (v -> Ax -> A^2x ->.....-> A^nv)
3) Keeps normalize to prevent from overflow
4) The vector converges to the fixed direction(dominant eigenvector)

```python
A = np.array([[2,1],[1,2]])
eig_val,eig_vec = np.linalg.eig(A)
x = np.array([3,4])

for _ in range(20):
    x = A@x/np.linalg.norm(A@x)

eig_val
```

# Connecting Eigendecomposition to SVD
It works for any matrices A E R^m*n.

A = U S V^T

V^T -> Input rotate

S -> Diagonal matrix of singular value

U -> Output rotate

```python
A = np.array([[1,3],[4,6]]) #for 2D

# # SVD Calculating data
U,S,VT = np.linalg.svd(A)
# # Diagonalized matrix
S = np.diag(S)

# Initially transformation with standard basis vector
basis_x = np.eye(2)
Ax = A @ basis_x

# # origin
origin = np.zeros(A.shape)

# # Calculating SVD (Transformation)
res = U @ S @ VT # reconstruct original matrix

# # Calculating Eigenvalue (left singular vectors)
eig_val,eig_vec = np.linalg.eig(A @ A.T)
left_val = np.sqrt(eig_val)

# Calculating Eigenvalue (right singular vectors)
eig_val,eig_vec = np.linalg.eig(A.T @ A)
right_val = np.sqrt(eig_val)

# # Outputs
print(f"Eigenvalue of left singular vectors U = {left_val}")
print(f"Eigenvalue of right singular vectors V = {right_val}")
print(f"Diagonalized  the singular value of a matrix=\n {S}")
print(f"Re-build originla matrix A =\n {res}")
```

```python
# A = np.array([[3,1,5,],[2,1,6],[5,3,2]]) # for 3D
A = np.random.randint(1,6, size=(3,3))
# SVD Calculating data
u,s,v = np.linalg.svd(A)
# Diagonalized matrix
s = np.diag(s)

# Initially transformation with standard basis vector
basis_x = np.eye(3)
Ax = A @ basis_x

# origin
origin = np.zeros(A.shape)

# Calculating SVD (Transformation)
res = u @ s @ v # reconstruct original matrix

# Calculating Eigenvalue (left singular vectors)
eig_val,eig_vec = np.linalg.eig(A @ A.T)
left_val = np.sqrt(eig_val)

# Calculating Eigenvalue (right singular vectors)
eig_val,eig_vec = np.linalg.eig(A.T @ A)
right_val = np.sqrt(eig_val)

# Outputs
print(f"Eigenvalue of left singular vectors U = {left_val}")
print(f"Eigenvalue of right singular vectors V = {right_val}")
print(f"Diagonalized of singular value of a matrix  =\n {s}")
print(f"Original matrix =\n {A}")
print(f"SVD result =\n {res}")
```
