# Vector addition

What is vector?

-> Vector is defined as under the operation of summation and scalar multiplication. (Theory)

-> Vector is an arrow which pointing in the space. (Geometry)

-> Vector is ordered list of numbers. (Computation)

```python
v = np.array(([2],[3]))
origin = np.zeros(2)

# vector scaling 

x = 2 #scalar
new_v = x*v
new_v
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *new_v, angles='xy', color='g', scale_units='xy', scale=1)

ax.grid()

ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_aspect('equal')

# arrow pointing
ax.text(v[0][0], v[1][0], 'original vector', color='k')
ax.text(new_v[0][0], new_v[1][0], 'scaled vector', color='r')

plt.show()
```

# Properties of vector
Distributivity, Associativity(Outer operation)

```python
A = np.random.randint(1,9, size=((3,3)))

v1 = A[:, 0]
v2 = A[:, 1]
v3 = A[:, 2]
V = np.asarray([v1,v2,v3])

K = 5 # where K -> scalar

# Distributivity
# K * (v1 + v2 + v3) == K*v1 + K*v2 + K*v3

# Associativity
# K * (v1 * v2 * v3) == (v1 * v2 * v3) * K
```

# Permutation Matrix

```python
I = np.eye(3, dtype=int)

# Swap rows
I[0] = I[0][::-1]
I[1] = I[1][::-1]
I[2] = I[2][::-1]

# Direct reverse Identity matrix
I = np.eye(3, dtype=int)[::-1]

# P = I

# OR 
P = np.zeros((3,3), dtype=int)
P[0,-1] = 1
P[1, 1] = 1
P[2, 0] = 1

print(P)
```

# Row exchange matrix

```python
A = np.array([[3,1,5],
              [4,2,1],
              [5,2,3]])

print(f"Before =\n {A}")

A_ex = P @ A

print(f"\nAfter =\n {A_ex}")
```

# Graphically

# In 2D

```python
A = np.array([[3,2],[4,5]])
P = np.zeros(A.shape, dtype=int)
P[0,-1] = 1
P[1, 0] = 1
A_ex = P @ A
origin = np.zeros(A.shape)

fig, ax = plt.subplots()

ax.quiver(*origin, *A, angles='xy', scale_units='xy', scale=1, color='k', label="Original Matrix")
ax.quiver(*origin, *A_ex, angles='xy', scale_units='xy', scale=1, color='r', label='Row Exchange Matrix')

ax.grid()
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.set_aspect('equal')
ax.legend()

plt.show()
```

# In 3D

```python
origin = np.zeros((3,3))
A = np.array([[3,1,5],
              [4,2,1],
              [5,2,3]])
P = np.zeros(A.shape, dtype=int)
P[0,-1] = 1
P[1, 1] = 1
P[2, 0] = 1

A_ex = P @ A
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.quiver3D(*origin, *A)
ax.quiver3D(*origin, *A_ex)


ax.set_xlim3d(-6,6)
ax.set_ylim3d(-6,6)
ax.set_zlim3d(-6,6)
ax.set_aspect('equal')

plt.show()
```
