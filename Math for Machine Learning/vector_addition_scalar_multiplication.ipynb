```python
import numpy as np
import matplotlib.pyplot as plt
```

# Vector addition

What is vector?

-> Vector is defined as under the operation of summation and scalar multiplication. (Theory)

-> Vector is an arrow which pointing in the space. (Geometry)

-> Vector is ordered list of numbers. (Computation)

```python
v = np.array(([2],[3]))
origin = np.zeros(2)

# vector scaling 

x = 2 #scalar
new_v = x*v
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *new_v, angles='xy', color='g', scale_units='xy', scale=1)

ax.grid()

ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_aspect('equal')

# arrow pointing
ax.text(v[0][0], v[1][0], 'original vector', color='k')
ax.text(new_v[0][0], new_v[1][0], 'scaled vector', color='r')

plt.show()
```

```python
v = np.array([[2],[3]])
w = np.array([[6],[1]])

origin = np.zeros(2)
sum_v_w = np.array([[4],[-2]])
sum_v_w
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1)
ax.quiver(*origin, *w, angles='xy', color='r', scale_units='xy', scale=1)
ax.quiver(*v, *sum_v_w, angles='xy', color='g', scale_units='xy', scale=1)

ax.grid()

ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_aspect('equal')

# arrow text
ax.text(v[0][0],v[1][0], 'v')
ax.text(w[0][0],w[1][0], 'w', color='r')
ax.text(sum_v_w[0][0],-sum_v_w[1][0], 'v+w', color='g')

plt.show()
```
