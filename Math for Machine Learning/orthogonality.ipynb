```python
import numpy as np
import matplotlib.pyplot as plt
```

# Orthogonal vector
dot produt of two vetors is 0.(a and b both are perpendicular to each other)


```python
v1 = np.array([3,2])
v2= np.array([4,-6])
origin = np.zeros(2)

np.dot(v1,v2)
```

```python
plt.quiver(*origin,*v1, angles='xy', scale_units='xy',scale=1)
plt.quiver(*origin,*v2, angles='xy', scale_units='xy',scale=1)

plt.grid()
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```

# Cosine similarity


```python
a = np.array([2,1,3])
b = np.array([4,3,1])
origin = np.zeros(2)
    
dot_product = np.dot(a,b)
norm1 = np.linalg.norm(a)
norm2 = np.linalg.norm(b)
if norm1 == 0 or norm2 == 0:
    print(False)
    
theta = (dot_product/(norm1 * norm2))
x = (theta*180)/np.pi

print(f"Angle between two vectors a and b is {np.round(x,2)}")
```

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.quiver(*origin, *a, angles='xy', scale_units="xy", scale=1)
ax.quiver(*origin, *b, color='g', angles='xy', scale_units="xy", scale=1)


ax.grid()

ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')

plt.show()
```
