```python
import numpy as np
import matplotlib.pyplot as plt
```

# Orthogonal vector
Inner/dot produt of two vetors is 0.(both vectors are perpendicular to each other)


```python
v1 = np.array([4, 4])
v2 = np.array([-4, 4])
origin = np.zeros(2)

alpha = np.dot(v1, v2)/(np.linalg.norm(v1) * np.linalg.norm(v2))
theta = (180 / np.pi) * np.pi/3 # radian convert into degree 

print(f"Angle = {theta}°")
print(f"Dot product v1 and v2 = {np.dot(v1,v2)}")
```

```python
plt.quiver(*origin,*v1, angles='xy', scale_units='xy',scale=1, color='r')
plt.quiver(*origin,*v2, angles='xy', scale_units='xy',scale=1, color='g')

plt.grid()
plt.xlim(-5,5)
plt.ylim(-5,5)


plt.show()
```

# Orthogonal matrix
A E R^n*n if columns are orthonormal, linear transformation (like rotation and reflection(flip)), preserve both distances and angles

orthonormal -> each column vector has unit lenght, i != j,  a1.a2 = 0

```python
A = np.array([[1, 0],
              [0, -1]])

v1 = A[:, 0]
v2 = A[:, 1]

dot_prod = np.dot(v1, v2)
alpha = dot_prod/np.linalg.norm(v1)*np.linalg.norm(v2)
angle = np.pi*alpha/180

print(f"Check orthogonal matrix:")
print(f"v1.v2 = {dot_prod}")
print(f"Euclidean vector space = {np.linalg.norm(v1)}")
print(f"Angle between two vectors v1 and v2 = {angle}")
```

# Cosine similarity


```python
a = np.array([2,1,3])
b = np.array([4,3,1])
origin = np.zeros(2)
    
dot_product = np.dot(a,b)
norm1 = np.linalg.norm(a)
norm2 = np.linalg.norm(b)
if norm1 == 0 or norm2 == 0:
    print(False)
    
theta = (dot_product/(norm1 * norm2))
x = (theta*180)/np.pi

print(f"Angle between two vectors a and b is {np.round(x,2)}")
```

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.quiver(*origin, *a, angles='xy', scale_units="xy", scale=1)
ax.quiver(*origin, *b, color='g', angles='xy', scale_units="xy", scale=1)


ax.grid()

ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect('equal')

plt.show()
```
