```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([2,3])
b = np.array([4,2])
origin = np.zeros(2)

# Calculating projection a on b
proj_a_on_b = np.dot(a,b)/np.linalg.norm(b)**2 * b

# Calculating error
error = a - proj_a_on_b

# output
print(f"Projection a on b = {proj_a_on_b}")
print(f"Error = {error}") 
# Error -> "How much component of a that is not represent on the direction of b."
# Or say; it is represent minimum possible error.
# error ⊥ b 
```

```python
fig, ax = plt.subplots(figsize=(8,7))

ax.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1, color='y')
ax.quiver(*origin, *b, angles='xy', scale_units='xy', scale=1, color='g')
ax.quiver(*origin, *proj_a_on_b, angles='xy', scale_units='xy', scale=1, color='b')
ax.plot([b[1],proj_a_on_b[0]], [a[1],proj_a_on_b[1]], 'r--')

# coordinates axes
ax.axhline(0, color='k')
ax.axvline(0, color='k')
ax.set_xlabel('y')
ax.set_ylabel('x')

# style
ax.grid()
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.set_aspect('equal')

# text
ax.text(a[0],a[1], s='a', color='k')
ax.text(b[0],b[1], s='b', color='g')
ax.text(1,0.2, s='proj_a_on_b', color='b')
ax.text(2.7,2, s='Error', color='r')

plt.show()
```
