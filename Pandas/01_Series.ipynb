# Series

A series is one dimensional array that containing array of data,( of any NumPy datatype) and an associsted arrary of data label called index.

This is a simplest series to make format of array of data. 

```python
series = pd.Series([90,89,100,23,45,900], index=(['a','b','c','d','e','f']), name=('Index'))
series.sort_values()
series[:-1]
# del series['a']
# series.pop('b')
series
```

# iloc -> ( full_name: Index location) using this method to access values of any specific index location.  In this case labeled keys index input

```python
#            row,  column
record.iloc[[0,1,2,5], 2] # this gives column index 2 -> all 0 to 3 indices values of 2-index
# record.iloc[[1,2]]  # this way to return 1 to 2 all DataFrame values
# record.iloc[1:5, [0,2]]                                                                                            
```

# loc is used to label based on indexing and access rows and columns 

```python
record.loc[[0,3], 'last_name'] # this way to return 0 and 3 indices rows values of last_name
record.loc[1] # this way to manually return each values of 1-index row with keys-labels

record.loc[[1,3,5,6], ['first_name','email']] # each specify index values of specify keys-labels

record.loc[1:6, 'first_name':'email']
```
