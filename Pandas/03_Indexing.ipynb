```python
import pandas as pd
import numpy as np
```

# index_col

This used to make column index of existing column of its index and column key.

```python
# data = pd.read_csv('gold_price_forecasting_dataset.csv', index_col='date')
# data
data = pd.read_csv('gold_price_forecasting_dataset.csv')
# data
data.high[:8].value_counts() #this return given key of its same values
# return each label keys in columns how many rows consists
data.count()
```

# Indexing

Indexing is same as like dict call keys its values

```python
dataframe = pd.DataFrame(data)
# low -> label key of column
dataframe.low #this is musch easier than below one
# both are used to access 
dataframe['low'][:5]
# total no. of columns
dataframe.columns

#Each first label columns data shows
dataframe.loc[0]
dataframe.loc[:5, ['low','high']]
```

# set_index()

It used to makes index of existing column of dataframe.
inplace is nothing but is used to makes also changes on original dataframe


```python
# dataframe.set_index('high')
# dataframe

dataframe.set_index('high', inplace=True)
dataframe
```

# reset_index()

```python
dataframe.reset_index(inplace=True)
dataframe
```

# sort_index (ascending or decending)

```python
# dataframe[:5].sort_index(ascending=True) # ascending order
dataframe.sort_index(ascending=False, inplace=True) # Decending order
dataframe
```

# Slicing

```python
dataframe.high[:5]
# reverse
dataframe.high[::-1]

dataframe.low[0:5:2] #even slicing
dataframe.high[1:5:2] #odd slicing

# Also can 
#             row  ,   columns
dataframe.loc[1:5, 'low':'volume']

dataframe.loc[[5,10,20,50]] #preview at instant any specific data

dataframe.iloc[1:5, 3:7]

dataframe.low 
```
