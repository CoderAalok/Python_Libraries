# Dict data into DataFrame

If data stored in dictionary, list or tuple then we can behaves them as it's method and attribute.

```python
import pandas as pd
import numpy as np
```

# Columns are specify in sequence order

```python
data = {'HeadPhone':[500,1000,1500,2000,4000,5000],
        'SmartWatch':[550,150,1000,500,600,8000],
        'EarPhone':[400,100,500,600,2000,1000]}

size = list(data.values())[0]
data['S.N.'] = list(range(1, len(size)+1))
df = pd.DataFrame(data, columns=['S.N.','EarPhone','HeadPhone','SmartWatch'], index=(range(1, len(size)+1)))
df
```

# Columns in DataFrame can be retrived as Series either by dict notation or by attribute

```python
# By dict notation
df['EarPhone']

# By attribute
df.HeadPhone
```

# Columns can be modified by assingment

# Nested format data stored

All nested/inner keys are becomes indices and all nested values are becomes a column.
Each first keys are labels in column

```python
data = {'ChatGPT':{'Version 1': 2.1, 'Version 2':2.5, 'Version 3':3.1, 'Version 4':3.6, 'Version 5':4.0, 'Version 6':5.0},
        'Claud Sonnet':{'Version 1':2.5, 'Version 2':3.1, 'Version 3':3.4, 'Version 4':3.6, 'Version 5':3.9, 'Version 6':4.5}}

df = pd.DataFrame(data)
df

# Also can transpose the data
# df.T


# Values attribute returns 2D ndarray
# df.values

# Keys attribute returns NDframe
# df.keys
```

# List of dicts

Each items are specify in rows
Each keys label in column and values are becomes a column.

```python
data = [{'Python':'3.14.0', 'C++':'4.8.9', 'Java':'4.05'}, {'Python':'3.14.1', 'C++':'4.9.9', 'Java':'4.8.5'}]
pd.DataFrame(data)

```
