# csv file makes in format  using csv module

```python
file = 'names.csv'
with open(file) as csv_file:
    read_file = csv.reader(csv_file)
    
    with open('format_data.csv', 'w')as f_data:
        write_file = csv.writer(f_data, delimiter='\t')
        
        for data in read_file:
            write_file.writerow(data)
```

# Using DictReader to make format csv file

```python
file = 'names.csv'
with open(file) as csv_file:
    read_file = csv.DictReader(csv_file)
    
    with open('format_data.csv', 'w')as f_data:
        col_names = ['first_name', 'last_name', 'email']
        write_file = csv.DictWriter(f_data, fieldnames=col_names, delimiter='\t')
        write_file.writeheader()
        
        for data in read_file:
            # del data['last_name']
            write_file.writerow(data)
            
```
