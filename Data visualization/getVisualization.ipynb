```python
y = np.linspace(0,2,100)
x = y

fig,ax = plt.subplots()

ax.plot(y,x, label='Linear', linewidth=2.5)
ax.plot(y,x**2, label='Quadratic', linewidth=2.5)
ax.plot(y,x**3, label='Cubic', linewidth=2.5)

# add axis
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend()
plt.show()
```

```python
y = np.linspace(0.1,3,100)
x = y

fig,ax = plt.subplots(layout='constrained')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.plot(y,log(x), label='Logarithm', linewidth=2.5)
ax.plot(y,e**x, label='Exponential', linewidth=2.5)

ax.legend()
plt.show()
```

```python
y = np.linspace(0,20,100)
x = y

fig,ax = plt.subplots(figsize=(10,6),layout='constrained')
  
ax.plot(y,sin(x), label='Sine function', marker='o')
ax.plot(y,cos(x), label='Cosine function', marker='o')

ax.legend()
plt.show()
```
