# Universal Function

A function that perform elementwise operation in ndarray. 

It takes one or more scalar and gives one or more scalar.

Many ufunc simple elementwise transformation like square root(sqrt), exponential(exp)

```python
arr = np.arange(10)
np.sqrt(arr)

np.round(np.exp(arr))
```

# randn()  or random.default_rng().standard_normal(()) -> newer version 

randn is random number generator that generating number from the standard normal distribution. 

That means:

mean=0

std=1

Distribution= Gaussian curve bell 

Normal distribution: In Mathematical convention behaviour for modeling noise, initillizing weight for neural networks, standradzing data, stable data distribution

```python
arr = np.random.randn(10)
np.mean(arr)
np.std(arr)

arr*10
```

# modf

It is built-in Python divmod is returns  fractional and integral parts of floating points of array.


```python
arr = np.random.randn(2,5)
np.modf(arr)

arr = np.random.default_rng().standard_normal((10))

plt.plot(arr)
plt.xlim(1,10)
plt.ylim(-2,2)
plt.show()
```
