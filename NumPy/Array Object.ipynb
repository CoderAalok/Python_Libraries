# dump

It is used to seriallization(pickling) means it is a process to converting Python object hierarchy to binary byte stream.
It help to save file and later reconstructed file like list, dictionary and Machine Learning Model.  
Uses Pickle to byte stream or file.

```python
import pickle
data = {"Model":321,
        'MonoDB':"SQL Lite",
        'Byte':'Bit',
        'Serialization':'Pickling'}

with open('file_binaray_data.pkl', 'wb')as file:
    pickle.dump(data, file, 2)

with open('file_binaray_data.pkl', 'rb')as file:
    loaded_file = pickle.load(file)

print(loaded_file)
```

# byteswap(bool)

It reverses the bytes order of each elements in an array.

Its matter in ML and systems:
>> Reading larger binary dataset of Machine model.
>> Loading the model weights from other architectures.

This matters in computer, Computers store numbers using endianness:

● Little-endian: Lowest/least significant byte stored first.

● Big-endian: Highest/most significant byte stored first.

● < : little endian

● > : big endian

● = : native

```python
data = array([1,2], dtype=np.int16)  # 2-bytes = 16-bits (1byte = 8bits -> 00000001) 
data.byteswap(False)
# data.dtype.byteorder
```

# copy()

It returns copy of original array. 

It preserved data type whether or not data is byteswapped.

```python
data = array(([4,2,2,90,32],
            [3,2,45,56,32],
            [12,2,33,3,4],
            [3,2,45,56,32],
            [12,2,33,3,4]), dtype=np.int32)

data_copy = data.copy()
data_copy
# data_copy[2] = 900
# data_copy.dtype
```

# view()
It reinterprets the same memory bits as different data type  (like int, float, uint8, uint16, uint32, uint64, bool).

That matters when reading binary raw data, hardware data stream, image processing, network packets.

```python
# data_copy.view(dtype=np.int64)  # argument as data type desicriptor gives -> new array same as self
data_copy.view(np.single).shape
data_copy.view(np.ubyte)  # returns an array with how many bytes each element consists 
```

# transpose()

```python
arr = array(([2,3,4,5],
             [3,2,3,5]))

arr.transpose()

arr.transpose() == arr.T

arr.transpose().shape == arr.shape[::-1] 
# arr.transpose().ndim
```

# swapaxes(axis1, axis2)

Swapaxes returns a view without copy of data  so when changes it affects on original data

```python
arr.swapaxes(0,1)
# arr.swapaxes(0,1) == arr.T

# arr = np.arange(30).reshape((2,3,5))
# arr.swapaxes(1,2)
# # arr

arr = np.indices((4,5))
arr.swapaxes(1,2)[0][2:] = 9
arr

# arr.ndim
# arr
```

# flatten(order='C')

It is a method used to converting multi-deminsional to 1D.

Changes on flatten 1D array, no affects on original array.

```python
arr = np.arange(1,21).reshape(5,4)

arr_f = arr.flatten(order='C')

# arr.flatten(order='F')

# arr_copy = arr.copy()

# arr_copy[3:] = 0

# arr_copy

arr_f[4:] = 0
arr_f
```

# ravel(order='C')

This also converting multi-dimensional to 1D but it provide much better efficieny than flatten.

Changes on ravel 1D array, also affects on original array.

```python
arr = np.arange(10,20).reshape(5,2) 
arr_r = arr.ravel()
arr_r[4:] = -1
arr
```
