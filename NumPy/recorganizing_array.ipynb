```python
import numpy as np
```

```python
# Vertical stacking

X = np.array(([1,2], 
              [3,2]))
Y = np.array(([3,1],
              [3,4]))
print(np.vstack([X,Y])) #use both parentheses and square barcket

# Horizontal stacking
print(np.hstack([X,Y])) #use both parentheses and square barcket
```

# Matrix Multiplication

```python
np.dot(X,Y)
np.multiply(X,Y)
```

# Boardcasting

```python
a = np.arange(1,5)  # no. of rows diff = 5-1= 4
b = np.arange(4,9)  # no. of column diff = 9-4= 5
# 

print(a.shape)
```

# Adding new axis

```python
print(a[:,np.newaxis])
a[:,np.newaxis]*b

# print('\n',b[:, np.newaxis])
```
