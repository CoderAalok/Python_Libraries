# Problem 4 — Broadcasting + Axis Logic

```python
matrix = np.random.default_rng(12).random(size=(4,3))

# # 1.
mean = matrix.mean(axis=0)
sub_col = matrix-mean
print(sub_col)

# 2.
std = (matrix-matrix.mean(axis=1, keepdims=True))/matrix.std(axis=1, keepdims=True)
# std = std.reshape(std.size,1)
# div = matrix/std 
print(std)
                
# 3.
# i) Column means approximately 0.
col_approx = sub_col.mean(axis=0)
print(col_approx)

# ii) Row standard deviations are approximately 1. -> this requires row standardrization

row_approx = (matrix-matrix.mean(axis=1, keepdims=True))/matrix.std(axis=1, keepdims=True)
row_approx
```

# Problem 5 — Transpose + Axis Manipulation

```python
arr = np.arange(24).reshape((3,4,2))

# Swap axis 0 and 2.
res = arr.swapaxes(0,2)  # arr.shape -> 3*4*2
print(res)  # res.shape -> 2*4*3  ( swaped 0 and 2 axis where 1 remains same)

# Then reshape the result into a 2D array.
res_new = res.reshape(8,3)
print(res_new)
```

What shape do you get after swapping?

Ans: After swapping shape -> (2 * 4 * 3)


What shape after reshaping?

Ans: After reshaping  shape -> (8,3) [shape depends on reshaping]

# Problem 7 — Boolean Masking Strategy
Boolean masking stratrgy(way) to selecting or filtering elements using boolean condition.

```python
data = np.random.randn(1000)

mask = (data >= -1) & (data <= 1)
data[~mask] = 0

# Mean
mean = np.mean(data)
print(mean)

# Standard deviation
std = np.std(data)
print(std)

# Percentage of zeros (In general, we are calculating invalid data in dataset)
count_zeros = np.sum(data==0)
percentage = count_zeros * 100/len(data)
print(f'{percentage} %')
```
