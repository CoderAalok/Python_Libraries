# Axis

```python
# Axis: column wise combining if axis=0, row wise combining if axis=1
arr = np.array(([5,3,4],
                [1,4,5]))
# column-wise
print(arr.sum(axis=0)) 
# row-wise
print(arr.sum(axis=1))

arr.shape
arr.strides  
```
