# Indexing & Slicing

```python
arr = np.arange(10,20)
print(arr)
print(arr[:3])
print(arr[-3:])
print(arr[::-2])

mat = np.array(([3,1,2],
                [4,5,2],
                [5,2,1]))
# gives row
print(mat[1, ...])
# gives column
print(mat[...,2])
```

```python
arr = np.arange(20).reshape((4,5))
print(arr)
# Extract the 2nd row
print(arr[1:2])
# Extract the 3rd column
print(arr[[0,1,2,3], 2])
# Extract the sub-matrix of the first 2 rows and first 3 columns
# First 2 rows
print(arr[0:2])
# first 3 columns
print(arr[[0,1,2], 0:3])
# Replace all elements in the last row of a matrix with -1.
arr[-1] = -1
print(arr)
```

# Basic Array Operations

```python
arr1 = np.array(([1,2,3],
                [4,5,6]))
arr2 = np.array(([0,2,3],
                [5,5,3]))
# Add
print(arr1 + arr2)
# subtract
print(arr1 - arr2)
# multiply
print(arr1 * arr2)


```

# Given an array, square every element without using a loop.

```python
arr = np.array(([0,2,3],
                [5,5,3]))
print(np.square(arr))
```

#  Normalize an array so that all values lie between 0 and 1.

```python
arr = np.array([100,200,300,400,500])
normalize = (arr - min(arr))/(max(arr) - min(arr))
print(normalize)
```

# Flatten a 2D array

```python
# Flatten: convert the multi-dimension array into 1D
arr = np.array(([2,1,4],[4,5,6]))
# question no. 11
flatten = arr.ravel()
print(f"Before flatten:")
print(arr)
print(f"After flatten: {flatten}")

# question no. 11
# copy flatten
c_flatten = flatten.copy()
# or another way to copy data
c_flatten = flatten.flatten()
c_flatten[3] = 900
# result there is no changes in original because it just a copy not reference takes
# original result
print(flatten)
  
# copy result
print(c_flatten)

# # question no. 12
# reshape a matrix
re_shape = arr.reshape(6,1)
print(re_shape)
```

# Boolean Indexing

```python
arr = np.random.randint(180, size=20)
# random array 
print(arr)

# Mean
mean_value = np.mean(arr)
print(mean_value)

# sum of all number
print(np.sum(arr))

# Extract all values greater than mean value
print(arr > mean_value)
```

```python
# replace all negative values in an array with 0.
arr = np.random.randint(-50,50, size=20)
# arr.view()

print(arr)

positive = arr[arr>0]
print(f"Positive: {positive}")
print(positive.shape)

negative = arr[arr<0]
print(f"Negative: {negative}")

print(negative.shape)
negative[:] = 0
# after replace with 0
print(negative)

count_greater = arr[arr>10]
print(count_greater)

count_less = arr[arr<10]
print(count_less)
```

# Aggregations

```python
# Aggregation is nothing is just an operations that used to combining all elements of an array into single value.
# operations like: sum, argmax, argmin, max, min
arr = np.array(([2,3,1],
                [1,4,5],
                [6,5,0]))
print(arr.shape)
# row-wise sum
row_wise_sum = arr.sum(axis=1)
print(row_wise_sum)
print(row_wise_sum.shape)

# column-wise sum
col_wise_sum = arr.sum(axis=0)
print(col_wise_sum)
print(col_wise_sum.shape)

# row-wise mean calculate
row_wise_mean = arr.mean(axis=1)
print(row_wise_mean)

# column-wise mean calculate
column_wise_mean = arr.mean(axis=0)
print(column_wise_mean)


# Find maximum value
max_val = np.max(arr)
print(max_val)

# Index of maximum value
ind_max_val = np.argmax(arr)
print(ind_max_val)

# Index of maximum value
ind_min_val = np.argmin(arr)
print(ind_min_val)


# Compute the standard deviation = sum(x-mean)^2/n-1 of an array.

standard_deviation = np.std(arr)
print(standard_deviation)

```
