# Student Performance Dataset (Data Representation)

```python
# students marks of five subjects (Math, Physics, Chemistry, Computer, English)
matrix = np.array((
                [80,75,90,75,95],
                [60,70,65,69,80],
                [90,88,85,65,86]))

print(matrix)
# Rows represent what?
# Each row represents a student marks for all subject.

# Columns represent what?
# Each column represents different sutudents same subject marks. 

print(matrix.shape)
print(matrix.ndim)
```

# Compute the average score per subject (column-wise mean)

```python
col_mean = matrix.mean(axis=0)
print(col_mean)
```

# Compute the average score per student (row-wise mean)

```python
row_mean = matrix.mean(axis=1)
print(row_mean)
```

# Feature Scaling (Normalization in Practice)

```python
data = np.array(([90,60,98,45],
                 [66,45,67,89],
                 [56,45,78,89],
                 [54,67,89,45]))

# Column-wise normalization
col_normalization = (data - data.min(axis=0))/(data.max(axis=0) - data.min(axis=0))
print(col_normalization)
```
