# compress(condition=, axis=None, out=None)

If condition true then given axis based on result gives. False are ignorance.

```python
arr = np.arange(5,25).reshape(5,4)
print(arr)
# pre-allocate conditions
cond = [True]*5
cond[1] = False
cond[4] = False

# output container creates to stored compressed result
output =  np.empty((4,4), dtype=arr.dtype)

np.compress(cond, arr, axis=0)

# output[[2,3], 1:3] = 0
# output

# arr
# # Rotated L -> Pattern 
# arr[[0,1,2,3,4],[0,0,0,0,0]] = 9
# arr[0] = 9

# arr[[1,2,3,4],[1,1,1,1]] = 8
# arr[1,1:] = 8

# arr[[2,3,4],[2,2,2]] = 7
# arr[2,2:] = 7

# arr[[3,4],[3,3]] = 6

# arr
```

# diagonal(offset=0, axis1=0, axis2=1)

offset: Basically, offset is shifts the diagonal starting point.

```python
arr = np.arange(1,26).reshape(5,5)
print(arr)
arr.diagonal(offset=0, axis1=-1, axis2=0)
```

# diag_indices(n) 
n -> shape of dimension of a square matrix


```python
v = np.array([[3,7]])

a = np.zeros((v.size,v.size))
# np.full_like(a,39) 

x,y = np.diag_indices(v.shape[-1])
a[x,y] = 1
a
```

# clip(min=, max=, out=None)

In general, we can apply limits in array.

Use case:

Trianing model

Image processing (for fixed pxiel valid range)

Overal, It use prevents from exceeding range.

```python
arr.clip(max=5)

np.clip(arr, min=1, max=100)

# stored clip result in output
output = np.empty((5,5), dtype=arr.dtype)

np.clip(arr, min=5, max=50, out=output)
output
```
