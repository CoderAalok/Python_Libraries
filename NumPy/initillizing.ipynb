Create matrices by rows and columns

```python
import numpy as np
#             row,col
print(np.zeros((3,3)))  # zero's matrix
print(np.ones([3,2])) #unit matrix
```

```python
# Random matrix creates
from numpy import shape

rand_matrix = np.random.randint(1,9,size=(4,3))
# rand_matrix = np.random.random_sample(size=(3,3))
print(rand_matrix)
```

```python
arr = np.array(([2,4,6,23,4,3,2],[2,1,6,23,4,3,2],[9,8,0,23,4,3,2]))
repeat_matrix = np.repeat(arr,2,axis=0)
print(repeat_matrix)
r = np.delete(repeat_matrix[-1],-1)
print("\n",r, repeat_matrix)
```

```python
# Get indexing/slicing of an element 
print(arr[2:,2]) #this return every 1 index elements (and slice)
print(arr[2,:-3,...]) # slice according to positional index
```
