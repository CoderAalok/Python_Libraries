Indexing on array

```python
import numpy as np
x = np.array(([3,5,3],[1,3,5],[9,8,7]))
y = np.array(([8,9,0],[8,9,7],[5,6,7]))
print(x*y)
print(np.prod(x))
# changing element from its position (n,m)
x[1,1] = 91
print(x)
```

Get silicing result from array

```python
print(x[1][::-1]) #rotating the given array
x[1,-1] = 90
print(x)
print(np.prod(x[-1]))

x = x[::-1]
print(x)
print(np.sum(x[-1]))
```
