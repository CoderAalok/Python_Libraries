```python
import numpy as np
```

# Contiguous memory layout

```python
# [1,2,3,4]  -> layout  
arr = np.array(([2,1,2,0],
                [1,23,5,6],
                [5,4,45,6],
                [0,3,4,6]))
arr.flags
```

# C_Style N-dimensional array

Indexing -> left increament first  (means column)

```python
print(arr)
arr.flags
```

# Fortran_Style N-dimensional array


Indexing ->right increments first (means row)

```python
arr1 = arr.T
print(arr1)
arr1.flags
```
