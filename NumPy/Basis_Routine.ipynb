# tensordot(a,b,axes=((-1,0)))

axes is called multiple dimensions

-1 -> shape of last dimension of a means first row of all columns 

we can also interchange axes .

0 -> shape of first dimension of b means every single row of all elements

when axes is 0 then expand the matrix (every single elements of a multiply with every single element of b)

when axes > 0 then collapse means values are combining along the axes.

Behind sense: last_dimension * first_dimension and sum it

```python
arr =np.arange(1,9).reshape(4,2)

np.tensordot(arr,arr1, axes=((-1,0)))  #every single row of elements of a multiplies with every single column of elements
# arr1.shape

np.tensordot(arr1,arr, axes=((0,-1)))  #every single column of elements of a multiplies with every single row of elements


# np.tensordot(arr,arr1, axes=0)  #  same like as outer
# np.tensordot(arr,arr1, axes=1)
```

# allclose(a,b, rtol=10^-8, atol=10^-5)

|a − b| < atol + rtol |b| .

So this is used to check toleraces.
Tolerance means if tiny threshold error found then  minimum changes required to continue the computation. 

```python
a = np.array([0,0])
b = np.array([0,0])
np.allclose(a,b, rtol=10**-8, atol=10**-5)
```
