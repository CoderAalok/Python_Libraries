```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.permutation(10)
np.random.uniform(4)
```

```python
distance = 100 #In Kilometer(Km) (in 1hr)
time = np.linspace(0.5,5,15) #In hour(hr)
speed = distance/time #Km/hr
```

```python
fig,ax = plt.subplots(figsize=(8,5))

ax.grid()
ax.plot(time, speed, marker='o')

# label
ax.set_xlabel("Time in hour(hr)")
ax.set_ylabel("Speed (Km/hr)")
plt.show()
```

# Random walk with step +1/-1

```python
position = 0
steps = 100

draws = np.random.randint(0,2, size=steps)
steps = np.where(draws>0, 1, -1)

walk = np.cumsum(steps)
walk
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(walk, marker='.')

plt.show()
```
