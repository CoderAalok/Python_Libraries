```python
import numpy as np
from numpy import array
```

# Section A: NumPy Basics & Operations

1. Array Creation & Data Types

```python
arr = np.arange(1,7).reshape(2,3)
print(arr)
print(arr.dtype)

# Convert it into float type
arr = np.arange(1,7, dtype=float).reshape(2,3)
print(arr)
print(arr.dtype)

# What is the type of the array elements?
# Ans: After convert the output array elements type is float.
```

2. Array Operation

```python
a = array(([1,2,3]))
b = array(([4,5,6]))

print(a+b) # adding of each element of a with b
print(a*b) # general multiplication of a and b
print(a @ b) # dot product of a and b

# What does a @ b represent mathematically?
# ans: It represent dot product of a and b.
```

3. Scalar Operations


```python
matrix = np.zeros((3,3), dtype=int)
matrix[[0,1,2], [0,1,2]] = 1 #used fancy indexing
print(matrix)

matrix = 5*matrix
print(matrix)

# What does each element represent after multiplication?
# ans: After multiplication by scalar diagonally becomes 5 and everywhere else 0.
```

# Section B: Indexing & Slicing

4. Basic Indexing

```python
arr = np.arange(16).reshape(4,4)
print(arr[0]) # first row
print(arr[:,-1]) # last column
print(arr[3,2]) # row 3 , column 2
```

5. Slicing

```python
arr[2:, 2:]
```

6. Boolean Indexing

```python
arr[arr>10]
```

7. Fancy Indexing

```python
# arr[0:3:2, [1,3]]
# arr[[0,2], 1:4:2]
arr[[0,2]][:, [1,3]]  #or
arr[np.ix_([0,2], [1,3])]
```

# Section C: Transpose & Axes

8. Transpose

```python
arr = np.arange(6).reshape(3,2)
print(arr) #before transpose
print(arr.shape) # -> (3,2)

arr.T # after transpose

# or
arr = arr.transpose()
print(arr)
print(arr.shape) # -> (2,3)
# What changes in shape?
# ans: After transpose row become column and column become row.
```

9. SwapAxes

```python
arr = np.arange(8).reshape(2,2,2)
print(arr)

print(arr.swapaxes(0,1))
arr.swapaxes(0,1) == arr.swapaxes(1,0)

# What is the new shape?
# ans: There is no changes on shape that is shape remain same in this case.

# self visualize
# matrix = np.indices((2,3,4))
# # len(matrix)
# # matrix.shape
# matrix.ndim
# print(matrix.swapaxes(0,1))
# matrix 
```

# Section D: Math/Linear Algebra Check

10. Vector Operations

```python
v1 = array([1,2,3])
v2 = array([4,5,6])

# Dot  product
print(np.vdot(v1,v2)) #-> 32

# Norm of v1
print(np.linalg.norm(v1))

# Projection of v1 onto v2
dot_product = np.vdot(v1,v2)

mag_v2 = v2 @ v2 #matrix multiplication
print(mag_v2)

mag_v2 = np.dot(v2,v2)
print(mag_v2)

print((dot_product/mag_v2) * v2) 
```

11. Matrix Operations

```python
a = np.arange(1,5).reshape(2,2)
b = np.arange(5,9).reshape(2,2)

print(a+b)
print(a*b)
print(a@b)
print(np.linalg.det(a))
```
