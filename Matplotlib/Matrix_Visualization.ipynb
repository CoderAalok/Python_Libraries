```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vec1 = np.array(([1],[1]))
vec2 = np.array(([-1,0],[0,1]))
vec = vec2 @ vec1
print(vec)
origin = np.zeros(vec.shape)
origin1 = np.zeros(vec1.shape)
```

```python
fig,ax = plt.subplots(figsize=(6,6), layout='constrained')

ax.quiver(*origin, *vec, scale=10)
ax.quiver(*origin1, *vec1, color=['r'], scale=10, )


ax.grid()
ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
plt.gca().set_aspect("equal")
plt.show()
```

# Linear Combination of two vectors

v = x*v1 + y*v2

```python
vec1 = np.array(([5],[3]))
vec2 = np.array(([2],[-3]))
x, y = -1,0.2

new_vec = x*vec1 + y*vec2
new_vec
```

```python
plt.figure(figsize=(8,6))
plt.quiver(*origin, *vec1, scale=1,  color=['r'], units='xy')
plt.quiver(*origin, *vec2, scale=1, color=['b'], units='xy')

plt.grid()
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.quiver(*origin, *new_vec, scale=1.1, color=['g'], units='xy')

plt.text(vec1[0][0], vec1[1][0], 'x', color='r')
plt.text(vec2[0][0], vec2[1][0], 'y', color='b')

plt.text(new_vec[0][0],new_vec[1][0], 'z', color='g')

plt.gca().set_aspect("equal")
plt.show()
```

# Linear Transformation

Transformation is a function that takes as an input vector and produced as an output vector(which is completely transformed)

Mapping input vector into output vector

```python
v = np.array(([-3],[2])) #input vector (-3i + 2j)
vec1 = np.array(([1,4],[1,3]))  # i-> [1,4], j-> [1,3]
new_vec = vec1.dot(v)  # output vector
print(new_vec)

origin = np.zeros(v.shape)
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.quiver(*origin, *v, color='r', scale=1, units='xy')
ax.quiver(*origin, *new_vec, color='b', scale=1, units='xy')

ax.grid()
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)

ax.text(v[0][0], v[1][0], 'Input Vector')
ax.text(new_vec[0][0], new_vec[1][0], 'Output Vector')

plt.show()
```
