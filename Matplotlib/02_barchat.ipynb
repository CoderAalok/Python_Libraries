# Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from typing import Counter
```

```python
years = np.arange(2011,2027)

python_data = [100000, 100050, 100100, 105000,
                200000, 200100, 400000, 500000, 
                550000, 600010,700050, 800500,
                850000, 950010, 990000, 1000000]

C_plus = [150000, 130000, 400000, 500000,
            600000, 700050,690000, 650000,
            800000, 750000, 790000, 890000,
            900000,900050, 900500,909000]

java = [150500,140000,130000,200000,
        250000, 290050,300000,350000,
        310000,320100,350500,390050,
        500000,540000, 800000,850000]

fig,ax = plt.subplots()


plt.style.use('fivethirtyeight')
fig.set_figwidth(12)

year_indices = np.arange(len(years))
width = 0.28

ax.bar(year_indices-width, python_data,width=0.4, label='Python')
ax.bar(year_indices, C_plus, width=width, label='C++')
ax.bar(year_indices+width, java, width=width,label='Java')

ax.set_title('Most popular language upto date')
ax.set_xlabel('Years')
ax.set_ylabel('Popularity Growth')

# ax.xaxis.set_minor_formatter
# ax.yaxis.set_major_locator

ax.set_xticks(ticks=year_indices, labels=years)

ax.legend()
```

# Growth of Programming Languages

```python
# Load csv file from directory
data = pd.read_csv('worth_languages.csv')

# count popularity
count_languages = Counter()

for language in data['LanguagesWorkedWith']:
    count_languages.update(language.split(';'))

# different programming languages
languages = []
#languages popularity
popularity = []

for count in count_languages.most_common(15):
    languages.append(count[0])
    popularity.append(count[1])

languages.reverse()
popularity.reverse()

fig,ax = plt.subplots()
#width custimize
fig.set_figwidth(10)
#hight custimize 
fig.set_figheight(8)

#add style in figure
plt.style.use('fivethirtyeight')

# plot-bar horizontally
ax.barh(languages, popularity)

ax.set_title("Most Popular Programming Languages")
ax.set_xlabel('No. of People Who Use')
ax.set_ylabel('Popularity of Programming Languages')

# display on figure
languages.reverse()
ax.legend(languages)

plt.show()
```
