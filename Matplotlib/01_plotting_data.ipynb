# Plotting data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data1 = np.array([3,4,5,6,7])
data2 = np.array([10,30,5,30,40])
fig,ax = plt.subplots()
ax.grid()

plt.title('Sample graph of data')
# ax.set_title('Data record')
ax.scatter(data1,data2) #marker show
ax.plot(data1, data2)
plt.show()
```

```python
from matplotlib.lines import lineMarkers, lineStyles
from sympy import Line2D


year = list(range(2005,2021))
# ABC school
student_data1 = [20,70,50, 50,5,
                 30,100,100,200,150,
                 90,50,250,225,500,550]
# XYZ school
student_data2 = [10,5,30,60,70,50,20,
                 200,300,200,400,405,
                 500,200,600,1000]


plt.style.use('fivethirtyeight')

plt.plot(year,student_data1, 'r-',linewidth=2, label='ABC School') # 'r^' -> this repersentation of color and marker of line instead of this also can color='r', marker='o',linestyle='^'
# plt.plot(year, student_data2, color='g', marker='o', linestyle='dashed',label='XYZ School')
plt.plot(year,student_data2, 'g-', linewidth=2, label='XYZ School')
 
plt.xlabel('← Years →')
plt.ylabel('← No. of students →')

plt.title('Admission growth')
# ax.yaxis.set_major_locator
# ax.xaxis.set_major_formatter
# plt.tight_layout()

plt.legend()
# plt.savefig('Growth analysis.png')

plt.show()
```

```python
years = np.arange(2010,2026)

python_data = [150050, 100050, 100100, 105000,
                200000, 200100, 400000, 500000, 
                550000, 600010,700050, 800500,
                850000, 950010, 999000, 1000000]

C_plus = [100000, 130000, 400000, 500000,
            600000, 700050,690000, 650000,
            800000, 750000, 790000, 890000,
            900000,900050, 900500,909000]

java = [100500,140000,130000,200000,
        250000, 290050,300000,350000,
        310000,320100,350500,390050,
        500000,540000, 600000,800000]

fig,ax = plt.subplots()

ax.plot(years, python_data, 'r-', linewidth=1.5, label='Python')
ax.plot(years, C_plus, 'b-', linewidth=1.5, label='C++')
ax.plot(years, java, 'g-', linewidth=1.5, label='Java')

ax.set_title('Programming Language Growth')
ax.set_xlabel('Years')
ax.set_ylabel('Popularity Growth')

ax.legend()
```

```python
csv_file = pd.read_csv('programming language trend over time.csv')

month = np.arange(1,13)

python = csv_file['Python']
java = csv_file['Java']
c_plus = csv_file['C++']

fig,ax = plt.subplots()
fig.set_figwidth(14)

ax.plot(month, python.head(12), label='Python')
ax.plot(month, java.head(12), label='Java')
ax.plot(month, c_plus.head(12), label='C++')
ax.legend()
```
