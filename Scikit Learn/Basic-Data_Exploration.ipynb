```python
import pandas as pd
```

```python
house_data = pd.read_csv('ramen-ratings.csv')
house_data.describe()
# house_data.columns
house_data.groupby('LotArea').YearBuilt.count()
```

# Extracting mean/average of house size

```python
avg_size = house_data['LotArea'].mean()
print(round(avg_size))
```

# How old is the newest house

```python
newest_house = round(2026 - house_data['YearBuilt'].max())  # current_year - built_year 
newest_house
```

# How much for minimum size of house

```python
idx = house_data['LotArea'].idxmin()

size = round(house_data['LotArea'][idx])

price = round(house_data['SalePrice'][idx])

print(f'Size: {size}')
print(f"Price: {price}")
```
